**What to remember:**

- You can create ad-hoc ensembles of classifiers using a scikit-learn `VotingEnsemble`
- Hard voting is a simple majority vote to ensemble the output from multiple classifiers.
- Soft voting uses a weighted voting scheme, where the confidence of each classifier is used to weight
  their contribution to the final classification.
  - To perform soft voting in scikit-learn, all classifiers must be able to produce the associated
    prediction probability of each of their predictions.
- Random forests and extra trees are a common type of ensemble that use a large number of decision trees and
  can often have good performance on certain types of ML data problems.
- An advanced technique for ensembling results is to build and train another classifier that takes the
  predictions of a set of input classifiers and is trained to produce a final prediction.
