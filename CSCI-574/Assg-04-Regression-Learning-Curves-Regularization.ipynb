```python
def last_nonzero_index(lst):
    for i, val in reversed(list(enumerate(lst))):
        if val != 0:
            return i+1
    return None
```

```python
for alpha in alpha_params:
    lasso_model = task5_lasso_model(x, y, alpha)
    model = lasso_model.named_steps['Lasso']
    num_non_zeroes = sum(1 for coef in model.coef_ if coef != 0.0)
    max_dimension = last_nonzero_index(model.coef_)
    print(f"{alpha}: {num_non_zeroes}/{max_dimension} {root_mean_squared_error(y, lasso_model.predict(x))}")
```

**What to remember:**

- `scikit-learn` has high level methods, like the `GridSearch` to help you train and compare the
  meta-parameters of many models to see how they affect performance.
- You usually should aim to first create model that is underfitting the data, then another that
  is overfitting it.
  - This helps give you some idea on the bounds of the loss function.
  - You can use learning curves, if you learn how to read them, to determine if models are under or
    orverfitting.
- Once you have a model that overfits, you can use different types of regularization to fight overfitting.
  - The goal will usually be to find the right amount that fights the overfitting (the learning curves converge)
  - And they converge at a loss level that is more like the overfit model than an underfit model.
